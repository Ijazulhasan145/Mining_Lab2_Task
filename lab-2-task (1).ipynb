```python
import numpy as np
data = pd.DataFrame({
    'PatientID': range(1, 11),
    'BloodSugar': [110, 190, None, 95, 300, 115, 87, 305, 92, None],
    'Cholesterol': [200, 185, 300, 280, None, 195, 230, None, 210, 250],
    'BP_Systolic': [120, 130, 250, 110, 160, 140, None, 180, None, 190],
    'BP_Diastolic': [80, 85, 100, 70, 120, 90, None, 95, 105, 110],
})
data
```

```python
data['BloodSugar'].loc[data['BloodSugar'].isna()] = data['BloodSugar'].median()
data['BP_Systolic'].loc[data['BP_Systolic'].isna()] = data['BP_Systolic'].median()


mean_blood_sugar = data['BloodSugar'].mean()
data['BloodSugar'] = data['BloodSugar'].where(data['BloodSugar'] <= 200, mean_blood_sugar)


data['Cholesterol'] = data['Cholesterol'].where(data['Cholesterol'] <= 250, 250)

data['BP_Diastolic'].loc[data['BP_Diastolic'].isna()] = data['BP_Diastolic'].median()

bp_values = data[['BP_Systolic', 'BP_Diastolic']].values
normalized_bp_values = normalize(bp_values, axis=1, norm='l2')

data[['BP_Systolic', 'BP_Diastolic']] = normalized_bp_values


data['Cholesterol_Above_220'] = data['Cholesterol'].apply(lambda x: 1 if x > 220 else 0)

print(data)
```

# Task3

```python
data = pd.DataFrame({
    'Day': range(1, 11),
    'Temperature': [25, None, 45, 39, 42, 38, None, 46, 50, None],
    'Humidity': [55, 75, None, 95, 105, 115, None, 65, 60, 85],
    'WindSpeed': [5, 15, None, 20, 35, 50, None, 55, 60, None]
})
data
```

```python
data['Temperature'].fillna(data['Temperature'].median(), inplace=True)
data['Humidity'].fillna(data['Humidity'].mean(), inplace=True)
data['WindSpeed'].fillna(0, inplace=True)

data['Humidity'] = data['Humidity'].where(data['Humidity'] <= 100, 100)


data['Temperature'] = data['Temperature'].apply(lambda x: 'High' if x > 40 else 'Normal')

scaler = StandardScaler()
data['WindSpeed'] = scaler.fit_transform(data[['WindSpeed']])

print(data)
```

# Task 4

```python
data = pd.DataFrame({
    'CustomerID': range(1, 11),
    'CallDuration': [15, None, 45, 10, 120, None, 35, 50, 20, 90],
    'DataUsage': [1.2, 1.5, 3.0, None, 2.5, None, 5.0, 7.5, None, 6.0],
    'SatisfactionScore': [4, 3, None, 5, None, 2, 1, 4, None, 5]
})
data
```

```python
data['CallDuration'].fillna(data['CallDuration'].median(), inplace=True)

mean_data_usage = data['DataUsage'].mean()
data['DataUsage'] = data['DataUsage'].where(data['DataUsage'] <= 5, mean_data_usage)

scaler = StandardScaler()
data['CallDuration'] = scaler.fit_transform(data[['CallDuration']])

normalizer = Normalizer(norm='l2')
data[['DataUsage']] = normalizer.fit_transform(data[['DataUsage']])

data['SatisfactionLabel'] = data['SatisfactionScore'].apply(lambda x: 'Low' if x < 3 else ('Medium' if x == 3 else 'High'))

print(data)
```

# Task 5

```python
data = pd.DataFrame({
    'ParticipantID': range(1, 11),
    'Height': [160, None, 170, 165, None, 180, None, 175, 160, 155],
    'Weight': [60, None, 80, 85, None, 95, 100, None, 70, 65],
    'Age': [25, None, 45, 50, None, 30, 35, None, 40, 20]
})
data
```

```python
avg_height_over_30 = data.loc[data['Age'] > 30, 'Height'].mean()
data['Height'].fillna(avg_height_over_30, inplace=True)

median_weight = data['Weight'].median()
data['Weight'] = data['Weight'].where(data['Weight'] <= 90, median_weight)

normalizer = Normalizer(norm='l2')
data[['Height', 'Weight']] = normalizer.fit_transform(data[['Height', 'Weight']])

data['Age'] = data['Age'].apply(lambda x: 'Older' if x > 30 else 'Younger')

data['WeightCategory'] = data['Weight'].apply(lambda x: 'Underweight' if x < 70 else ('Normal' if x <= 85 else 'Overweight'))

print(data)
```
